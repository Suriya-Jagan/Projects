### Importing Libraries / Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

### Loading the Dataset

```python
path = 'HR-Employee-Attrition.csv'
data = pd.read_csv(path)
df = pd.DataFrame(data)
pd.set_option('display.max_columns', None)
df.head()
```

```python
# Deleting Duplicates if any
df = df.drop_duplicates()
df.shape
```

There is no duplicates rows found in dataset

```python
# Deleting the similar value Columns
unique_values = df.nunique() == 1
cols_to_drop = unique_values[unique_values].index
df.drop(columns = cols_to_drop, inplace = True)
df.shape  # 3 unique value columns dropped
```

```python
# Deleting the unique value in a Column it would be useless for this analysis
df.EmployeeNumber.nunique() == len(df.EmployeeNumber)
df.drop(columns = 'EmployeeNumber', inplace = True)
```

```python
df.shape
```

```python
# Renaming Columns with correct Spaces:
df.columns
new_col = []
for column in df.columns:
    count = 0
    char = ""
    for i in column:
        if i.isupper():
            count +=1
            if count > 1:
                char += ' ' + i
            else:
                char += i
        else:
            char += i
    new_col.append(char)
df.columns = new_col
```

### Attrition count by gender using Bar Graph

```python
plt.figure(figsize = (5,4))
a = sns.countplot(x = 'Gender', hue = 'Attrition', data = df, palette = 'Set1')

for p in a.patches:
    height = p.get_height()
    a.annotate(format(height, '.0f'), (p.get_x() + p.get_width() / 2., height), 
               ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.title('Attrition Count by Gender')
plt.show()
```

### Attrition based on Total years of experience

```python
#Creating the new column to group the ages
df['Total Working Years'].max()
df['Total Working Years'].min()

bins = [0, 6, 11, 21, 31, 41]
labels = ['0-5', '6-10','11-20','21-30', '31-40']

#Creating a Age Group column based on Age 
df['Total Working Years Group'] = pd.cut(df['Total Working Years'], bins = bins, labels = labels, right = False)

plt.figure(figsize = (5, 4))
a = sns.countplot(x = 'Total Working Years Group', hue = 'Attrition', data = df, palette = 'Set1')

for p in a.patches:
    height = p.get_height()
    a.annotate(format(height, '.0f'), (p.get_x() + p.get_width() / 2, height), ha = 'center',
               va = 'center', xytext = (0, 5), textcoords = 'offset points')

plt.xticks(rotation = 25)
plt.xlabel('Total Working Years')
plt.ylabel('Number of Employees')
plt.show()
```

### Attrition by Education using Heatmap

```python
contingency_table = pd.crosstab(df['Education'], df['Attrition'])
plt.figure(figsize=(10, 6))

sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlGnBu')

plt.xlabel('Attrition')
plt.ylabel('Education Level')
plt.title('Heatmap of Attrition by Education Level and Age Group')
plt.xticks(rotation=0)

plt.show()
```

### Number of Companies Worked VS Attrition

```python
grouped = df.groupby(['Num Companies Worked', 'Attrition']).size().unstack()

ax = grouped.plot(kind='bar', stacked=True, figsize=(10, 6))

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()    
    ax.annotate(format(height,'.0f'), (x + width/2, y + height/2), ha='center', va='center')


plt.xlabel('Number of Companies Worked')
plt.ylabel('Count')
plt.title('Number of Companies Worked vs. Attrition')

plt.legend(title='Attrition')
plt.xticks(rotation=0)
plt.show()
```

### Job Satisfaction based on job role

```python
df[['Job Satisfaction', 'Job Role']]
group = df.groupby(['Job Role', 'Job Satisfaction']).size().unstack()

total_employees_per_role = df.groupby('Job Role')['Job Satisfaction'].count()

percentages = group.div(total_employees_per_role, axis=0) * 100

percentages.plot(kind='bar', stacked=True, colormap='Set2')
plt.xlabel('Job Role')
plt.ylabel('Percentage of Employees (%)')
plt.title('Job Satisfaction by Job Role (Percentages)')
plt.xticks(rotation=45)
plt.legend(title='Job Satisfaction Level')
plt.grid(axis='y')

plt.ylim(0, 120)

plt.show()
```
